<p style="font-family:Verdana; font-size: 26px; color: cyan"> Change Axis Labels, Set Title and Figure Size to Plots with Seaborn</p>

```python
# Import required libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Load data set
tips = sns.load_dataset( "tips" )
tips.head()
```

> Example 1: Customizing plot with axes object

```python
# We make use of the set_title(), set_xlabel(), and set_ylabel() functions 
# to change axis labels and set the title for a plot. 
# We can set the size of the text with size attribute. 

# Set figure size (width, height) in inches
fig, ax = plt.subplots(figsize = ( 5 , 3 ))

# Plot the scatterplot
sns.scatterplot( ax = ax , x = "total_bill" , y = "tip" , data = tips )

# Set label for x-axis
ax.set_xlabel( "Total Bill (USD)" , size = 12 )

# Set label for y-axis
ax.set_ylabel( "Tips (USD)" , size = 12 )

# Set title for plot
ax.set_title( "Bill vs Tips" , size = 24 )

# Display figure
plt.show()
```

> Example 2: Customizing scatter plot with pyplot object

```python
# We can also change the axis labels and set the plot title with the matplotlib.pyplot object
#  using xlabel(), ylabel() and title() functions


# Set figure size (width, height) in inches
plt.figure(figsize = ( 5 , 3 ))

# Plot scatterplot
sns.scatterplot( x = "total_bill" , y = "tip" , data = tips )

# Set label for x-axis
plt.xlabel( "Total Bill (USD)" , size = 12 )

# Set label for y-axis
plt.ylabel( "Tips (USD)" , size = 12 )

# Set title for figure
plt.title( "Bill vs Tips" , size = 24 )

# Display figure
plt.show()
```

> Example 3: Customizing multiple plots in the same figure

```python
# Seaborn's relplot function returns a FacetGrid object which is a figure-level object. 
# This object allows the convenient management of subplots. 
# Set figure size
sns.set_theme( rc = {'figure.figsize' : ( 20, 20 ), 
               'axes.labelsize' : 12 })

# Plot scatter plot
g = sns.relplot(data = tips , x = "total_bill" , 
                y = "tip" , col = "time" , 
                hue = "day" , style = "day" , 
                kind = "scatter" )

# Title for the complete figure
g.figure.suptitle("Tips by time of day" ,
               fontsize = 'x-large' , 
               fontweight = 'bold' )

# Adjust subplots so that titles don't overlap
g.figure.subplots_adjust( top = 0.85 )

# Set x-axis and y-axis labels
g.set_axis_labels( "Tip" , "Total Bill (USD)" )

# Display the figure
plt.show()
```
