> **set_style()** method is used to set the aesthetic of the plot.
>> set_style(style=None, rc=None)

* darkgrid
* whitegrid
* dark

```python
# loading dataset 
data = sns.load_dataset("iris") 

# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 

# changing the theme to dark
sns.set_style("dark")
plt.show()
```

> Removal of Spines

```python
# loading dataset 
data = sns.load_dataset("iris") 

# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 

# Remove the top and right spines from plot(s).
sns.despine()
plt.show()
```

> Setting the Style Temporarily

```python
# loading dataset 
data = sns.load_dataset("iris") 

# Setting the scale of the plot
sns.set_context("notebook")

def plot():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data) 

# (function) def axes_style(
#     style: dict[str, Any] | Literal['white', 'dark', 'whitegrid', 'darkgrid', 'ticks'] | None = None,
#     rc: dict[str, Any] | None = None
# )
    
with sns.axes_style('darkgrid'):
    
    # Adding the subplot
    plt.subplot(211)
    plot()

# Setting the scale of the plot
sns.set_context("notebook")
    
plt.subplot(212)
plot()
```

<p style="font-family:Verdana; font-size: 22px; color: orange"> Multiple plots with Seaborn</p>

> Example 1: Using add_axes() method

```python
def graph():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Creating a new figure with width = 5 inches
# and height = 4 inches
fig = plt.figure(figsize =(5, 4)) 

# Creating first axes for the figure
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 

# plotting the graph
graph()

# Creating second axes for the figure
ax2 = fig.add_axes([0.5, 0.5, 0.3, 0.3])
# def add_axes(
#     rect: tuple[float, float, float, float],
#     projection: str | None = ...,
#     polar: bool = ...,
#     **kwargs: Any
# )
# plotting the graph
graph()

plt.show()
```

> Example 2: Using subplot() method

```python
def graph():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Adding the subplot at the specified
# grid position
plt.subplot(121)
graph()

# Adding the subplot at the specified
# grid position
plt.subplot(122)
# Call signatures:

#    subplot(nrows, ncols, index, **kwargs)
#    subplot(pos, **kwargs)
#    subplot(**kwargs)
#    subplot(ax)
graph()

plt.show()
```

> Example 3: Using subplot2grid() method

```python
def graph():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# adding the subplots
axes1 = plt.subplot2grid (
  (7, 1), (0, 0), rowspan = 2,  colspan = 1) 
graph()
# (function) def subplot2grid(
#     shape: tuple[int, int],
#     loc: tuple[int, int],
#     rowspan: int = 1,
#     colspan: int = 1,
#     fig: Figure | None = None,
#     **kwargs: Any
# ) 
axes2 = plt.subplot2grid (
  (7, 1), (2, 0), rowspan = 2, colspan = 1) 
graph()
  
axes3 = plt.subplot2grid (
  (7, 1), (4, 0), rowspan = 2, colspan = 1)
graph()
```

<p style="font-family:Verdana; font-size: 22px; color: orange"> Seaborn also provides some functions for plotting multiple plots.</p>

> Method 1: Using FacetGrid() method
>> Multi-plot grid for plotting conditional relationships.

```python

# loading dataset
data = sns.load_dataset("iris")

plot = sns.FacetGrid(data, col="species")
plot.map(plt.plot, "sepal_width")

plt.show()
```

> Method 2: Using PairGrid() method
>> Initialize the plot figure and PairGrid object.

```python
# loading dataset
data = sns.load_dataset("flights")

plot = sns.PairGrid(data)
plot.map(plt.plot)

plt.show()
```

<p style="font-family:Verdana; font-size: 26px; color: magenta"> Creating Different Types of Plots</p>

<p style="font-family:Verdana; font-size: 22px; color: orange"> Relational Plots</p>

> Relplot()
>> seaborn.relplot(x=None, y=None, data=None, **kwargs) 

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

# creating the relplot
sns.relplot(x='sepal_width', y='species', data=data)

plt.show()
```

> Scatter Plot
>> seaborn.scatterplot(x=None, y=None, data=None, **kwargs)

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=data)
plt.show()
```

> Line Plot
>> seaborn.lineplot(x=None, y=None, data=None, **kwargs)

```python
sns.lineplot(x='sepal_length', y='species', data=data)
plt.show()
```

> Bar Plot
>> barplot([x, y, hue, data, order, hue_order, …])

```python
sns.barplot(x='species', y='sepal_length', data=data)
plt.show()
```

> Count Plot
>> countplot([x, y, hue, data, order, …])

```python
sns.countplot(x='species', data=data)
plt.show()
```

> Box Plot
>> boxplot([x, y, hue, data, order, hue_order, …])

```python
sns.boxplot(x='species', y='sepal_width', data=data)
plt.show()
```

> Violinplot
>> violinplot([x, y, hue, data, order, …])

```python
sns.violinplot(x='species', y='sepal_width', data=data)
plt.show()
```

> Stripplot
>> stripplot([x, y, hue, data, order, …])

```python
sns.color_palette("Paired")

sns.swarmplot(x='species', y='sepal_width', data=data)
plt.show()
```

<p style="font-family:Verdana; font-size: 22px; color: orange"> Regression Plots</p>

> lmplot() method can be understood as a function that basically creates a linear model plot.

```python
# loading dataset
data = sns.load_dataset("tips")

sns.lmplot(x='total_bill', y='tip', data=data)
plt.show()
```

> regplot() method is also similar to lmplot which creates linear regression model.

```python
sns.regplot(x='total_bill', y='tip', data=data)
plt.show()
```
