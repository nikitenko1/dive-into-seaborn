<p style="font-family:Verdana; font-size: 22px; color: cyan"> Python Seaborn Tutorial</p>

```python
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt  
```

```python
# loading dataset 
data = sns.load_dataset("iris") 

# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
```

<p style="font-family:Verdana; font-size: 22px; color: orange"> Using Seaborn with Matplotlib</p>

> Example 1: 
>> We will be using the above example and will add the title to the plot using the Matplotlib.

```python
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 

# setting the title using Matplotlib
plt.title('Title using Matplotlib Function')
```

> Example 2: 
>> Setting the xlim and ylim

```python
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 


# setting the x limit of the plot
plt.xlim(5)

plt.show()
```

<p style="font-family:Verdana; font-size: 22px; color: orange"> Customizing Seaborn Plots</p>

> **set_style()** method is used to set the aesthetic of the plot.
>> set_style(style=None, rc=None)

* darkgrid
* whitegrid
* dark
* white
* ticks

```python
# loading dataset 
data = sns.load_dataset("iris") 

# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 

# changing the theme to dark
sns.set_style("dark")
plt.show()
```

> Removal of Spines
>> sns.despine(left = True)

```python
# loading dataset 
data = sns.load_dataset("iris") 

# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 

# Removing the spines
sns.despine()
plt.show()
```
